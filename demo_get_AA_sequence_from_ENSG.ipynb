```python
import requests, sys
 
server = "https://rest.ensembl.org"
ext = "/sequence/id/ENSP00000256442?multiple_sequences=1;type=protein"
 
r = requests.get(server+ext, headers={ "Content-Type" : "text/x-seqxml+xml"})
 
if not r.ok:
    r.raise_for_status()
    sys.exit()
 
 
print(r.text)
```
